```python
import csv

with open ("car.data", "r") as f:
    car_data= csv.reader(f, delimiter= '\n')
    query = """insert into cardataset7.car7(buying, maint, doors, persons, lug_boot, safety, class) values (%s, %s, %s, %s,%s, %s, %s)"""
    for i in car_data:
        split = str(i[0]).split(",")
        cursor.execute(query, split)
mydb.commit()
```
